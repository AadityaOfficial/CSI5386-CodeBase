```python
def fetch_poems(path):
    poems_arr = []
    for p in os.listdir(path):
        poem_path = os.path.join(path, p)
        f = open(poem_path,encoding='utf-8')
        poem_text = ""
        for i in f:
            poem_text += i.strip() + "\t"
        poems_arr.append(poem_text)
        f.close()
    return poems_arr
```

```python
from nltk import word_tokenize

def tokenize_words(upd_poem_cat):
    list_tokenize = []

    for poems in upd_poem_cat:
        tokenized_wrds = word_tokenize(poems)
        list_tokenize.append(tokenized_wrds)
    
    return list_tokenize


updated_p_anger = tokenize_words(updated_p_anger)
updated_p_animal = tokenize_words(updated_p_animal)
updated_p_children = tokenize_words(updated_p_children)
updated_p_father = tokenize_words(updated_p_father)
updated_p_friend = tokenize_words(updated_p_friend)
updated_p_river = tokenize_words(updated_p_river)
updated_p_romance = tokenize_words(updated_p_romance)
updated_p_water = tokenize_words(updated_p_water)
updated_p_winter = tokenize_words(updated_p_winter)
updated_p_sky = tokenize_words(updated_p_sky)

updated_p_anger
```

## Remove Stopwords ##

```python
from nltk.corpus import stopwords
stop_words_eng = stopwords.words('english')
print(stop_words_eng)
```

```python
def remove_stopwords(upd_poem_cat):
    filt_wrd_list=[]
    
    for words_all in upd_poem_cat:
        for lmn in range(len(words_all)):
            words_all[lmn] = words_all[lmn].lower()

        filtered_words = [word for word in words_all if word not in stop_words_eng]    
        filt_wrd_list.append(filtered_words)
    
    return filt_wrd_list


updated_stp_p_anger = remove_stopwords(updated_p_anger)
updated_stp_p_animal = remove_stopwords(updated_p_animal)
updated_stp_p_children = remove_stopwords(updated_p_children)
updated_stp_p_father = remove_stopwords(updated_p_father)
updated_stp_p_friend = remove_stopwords(updated_p_friend)
updated_stp_p_river = remove_stopwords(updated_p_river)
updated_stp_p_romance = remove_stopwords(updated_p_romance)
updated_stp_p_water = remove_stopwords(updated_p_water)
updated_stp_p_winter = remove_stopwords(updated_p_winter)
updated_stp_p_sky = remove_stopwords(updated_p_sky)

updated_stp_p_anger
```

```python
def vectors(nlp, data):
    vectors = []
    for row in data:
        row = str(row)
        doc = nlp(row)
        vectors.append(doc.vector)
    return np.array(vectors)
```

```python
import spacy
nlp = spacy.load('en_core_web_md')
train_vectors = vectors(nlp, X_train)
```

```python
test_vectors = vectors(nlp, X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix,classification_report
def evaluate(y_true, y_pred):
    # print evaluation results for model
    # print confusion matrix
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred))
    # accuracy
    print("Accuracy: ", accuracy_score(y_true, y_pred))
    # precision
    print("Precision: ", precision_score(y_true, y_pred, average = 'weighted'))
    # recall
    print("Recall ", recall_score(y_true, y_pred, average = 'weighted'))
    # f-measure
    print("F1: ", f1_score(y_true, y_pred, average = 'weighted'))
    print("F1: ", classification_report(y_true, y_pred))
```

```python
evaluate(y_test, y_pred_dt)
```
